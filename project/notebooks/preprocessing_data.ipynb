## Exploring the data

### About dataset
This dataset provides a comprehensive collection of attributes for various animal species worldwide. It includes detailed information such as Animal name, Height, Weight, Color, Lifespan, Diet, Habitat, Predators, Average and Top Speed, Countries Found, Conservation Status, Family, Gestation Period, Social Structure, and Offspring per Birth. These data points allow researchers and enthusiasts to analyze animal characteristics, study habitats and behaviors, explore dietary patterns, assess conservation statuses, and conduct ecological and wildlife studies effectively.

The primary task associated with this dataset involves data cleaning and classification of animals into three categories based on their diet: Omnivore, Carnivore, and Herbivore.

#### Dataset Glossary (Column-wise)
- Animal - Name of the animal.
- Height (cm) - Height range in centimeters for the animal.
- Weight (kg) - Weight range in kilograms for the animal.
- Color - Common colors associated with the animal's appearance.
- Lifespan (years) - Average lifespan of the animal in years.
- Diet - Type of diet the animal primarily follows (e.g., Carnivore, Herbivore).
- Habitat - Typical habitat or environment where the animal is found.
- Predators - Natural enemies or organisms that prey on the animal.
- Average Speed (km/h) - The average speed range the animal can achieve in kilometers per hour.
- Countries Found - Countries or regions where the animal is commonly found.
- Conservation Status - The conservation status of the animal as per relevant conservation organizations.
- Family - Taxonomic family the animal belongs to.
- Gestation Period (days) - Range of days representing the gestation or pregnancy period of the animal.
- Top Speed (km/h) - The maximum speed the animal can achieve in kilometers per hour.
- Social Structure - Information about the social behavior or structure of the animal (e.g., Solitary, Group-based).
- Offspring per Birth - The typical number of offspring born per birth or reproduction event for the animal.

## Data cleaning

### Duplicates values

First, let's address duplicate data by retaining entries with fewer NaN values.

```python
df['NAN sum' ] = df.isna().sum(axis=1)

df = df.sort_values(by=['Animal', 'NAN sum'])
df = df.drop_duplicates(subset=['Animal'], keep='first')
df = df.drop(columns=['NAN sum']).reset_index(drop=True)
```

```python
df.isnull().sum()
```

Next, we'll drop rows where more than 30 percent of the data is missing.

```python
nan_percentage = df.isna().mean(axis=1) * 100

threshold = 30.0
rows_to_drop = df[nan_percentage > threshold]

df = df.drop(rows_to_drop.index)
```

Average speed and top speed are strongly correlated; therefore, to simplify our analysis, we will fill missing values in the average speed data using the top speed values and subsequently drop the top speed column.

```python
df['Average Speed (km/h)'] = df['Average Speed (km/h)'].fillna(df['Top Speed (km/h)'])
df = df.drop(columns=['Top Speed (km/h)'])
df['Average Speed (km/h)'] = df['Average Speed (km/h)'].replace('Varies', np.nan)
```

### Target data

In the target column ('Diet'), there are 16 unique values, which we have consolidated into three main categories: Omnivore, Carnivore, and Herbivore.

```python
df['Diet'].value_counts()
```

```python
def map_diet(diet):
    diet = diet.lower()
    if 'omnivore' in diet or 'nectar, insects' in diet:
        return 'Omnivore'
    elif 'carnivore' in diet or 'piscivore' in diet or 'scavenger' in diet or 'insectivore' in diet:
        return 'Carnivore'
    elif 'herbivore' in diet or 'filter feeder' in diet:
        return 'Herbivore'
    
    else:
        return diet
    
df['Diet'] = df['Diet'].apply(map_diet)
```

### Incorrect values

Let's preprocess the data by removing prefixes such as 'Ups to' and 'usully', and converting string values to numerical by averaging the data.

```python
def process_range(range_str):
    multi = 1
    if type(range_str) is float or type(range_str) is int:
        return range_str
    elif range_str == '6-8 months':
        return 7/12
    elif range_str == '1-2 weeks':
        return 2 / 28
    elif range_str == '260-330 million years':
        return 5
    elif range_str == '1-4 (usually 2-3)':
        return 3
    elif range_str == 'Hundreds':
        return 1e2
    elif range_str == 'Thousands':
        return 1e3
    elif 'months' in range_str:
        multi = 30
    elif 'tons' in range_str:
        multi = 1e3
    elif 'weeks' in range_str:
        multi = 7
    
    
    range_str = re.sub(r'[^0-9\-.]', '', range_str)
    
    if '-' in range_str:
        num1, num2 = map(float, range_str.split('-'))
        return (num1 + num2) / 2
    else:
        return float(range_str) * multi

columns_to_apply =['Height (cm)', 'Weight (kg)', 'Lifespan (years)', 'Average Speed (km/h)', 'Gestation Period (days)', 'Offspring per Birth']
for col in columns_to_apply:
    df[col] = df[col].apply(process_range)
```
